# Correlation Matrix

```python
from statsmodels.graphics.correlation import plot_corr
```

```python
corr = df_selected.corr()
corr
```

```python
#Plotting the heatmap
fig, ax = plt.subplots(figsize=(20, 15))
sns.heatmap(corr, annot=True, fmt='0.2f', linewidths=1, cmap='Blues')
plt.title('Heatmap of Correlation Values')
plt.xticks(size=13)
plt.yticks(size=13)
plt.show()
```

- s9 and s14 are highly correlated with each other
- There is almost zero correlation between setting1 and TTF
- There is almost zero correlation between setting2 and TTF
- s11 and s12 are highly correlated with each other

Therefore we don't need them as our features

```python
df_dropped_from_corr = df_selected.drop(['s14', 's12','setting1', 'setting2'],1)
df_dropped_from_corr.head()
```

# Variance Inflation Factor (VIF)

```python
vif = []
for c in df_dropped_from_corr.columns[:-1]:
    X = df_dropped_from_corr.drop(['ttf'] + [c], axis=1)
    y = df_dropped_from_corr[c]
    X = sm.add_constant(X)
    est = sm.OLS(y, X).fit()
    r2 = est.rsquared
    vif.append((c, 1/(1-r2)))
vif
```

High VIF in s11 and s13 indicating strong multicollinearity. Therefore we remove these features

```python
final_df = df_selected.drop(['s11','s13'],1)
final_df.head()
```

```python
final_df.to_csv('datasets/predictive_maintenance_preprocessed.csv', index=False)
```
